```python
# First i import these libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Justin/Desktop/Anime.csv")
df
```

```python
# shows top 5 rows
df.head()
```

```python
# how many rows and columns in my dataset
df.shape
```

```python
# columns names
df.columns
```

```python
# checked datatypes
df.info()
```

```python
# tells which columns have missing values
df.isnull().sum()
```

```python
# handle missing values
df['Rating'].fillna(df['Rating'].mean(), inplace=True)
df['Episodes'].fillna(0, inplace=True)
```

```python
# Data Cleaning
# Convert Episodes to numeric
df['Episodes'] = pd.to_numeric(df['Episodes'], errors='coerce') #means non numeric values replace to nan
```

```python
# Remove duplicates
df.drop_duplicates(inplace=True)
```

```python
# Basic Statistics(gives mean,median and mode)
df.describe()
```

```python
#charts Analysis 
# Rating distribution
sns.histplot(df['Rating'], bins=20,label='Rating distribution',color='red')
plt.title('Histrogram  Chart')
plt.legend()
plt.show()
```

Top Anime Types

```python
df['Type'].value_counts().plot(kind='bar',color='green',label='Anime Types')
plt.title('BAR Chart')
plt.xlabel('Anime Types')
plt.ylabel('Counts')
plt.legend()
plt.show()
#Shows most common anime types
```

Rating vs Episodes

```python
sns.scatterplot(x='Episodes', y='Rating', data=df,hue='Type') # hue creates legend using parameter
plt.title('Relationship Scatterplot Chart')
plt.legend()
plt.legend(loc='lower right') 
plt.show()
# We check relationship between episodes and rating
```

Type vs Rating

```python
sns.boxplot(x='Type', y='Rating',color='cyan', data=df)
plt.legend()
plt.show()
#Compare ratings across different anime types
```

```python
#box shows 50 % of data
# bottom of box 25 percentile
# top of box 75th percentile
# box shows where most rating lie of each anime type
# line inside box means basically middle value(typical of that anime type)
# dots means very high or very low rating ( outliers )
```

Top Analysis for animes

```python
# Top 10 Anime by Rating
top_10 = df.sort_values(by='Rating', ascending=False).head(10)
top_10[['Name','Rating']]
```

Top Studios most popular

```python
df['Studio'].value_counts().head(10).plot(kind='bar',color='pink')
plt.title('Most Popular Studios')
plt.xlabel('Studio Names')
plt.ylabel('Counts')
plt.show()
```

```python
# Correlation Heatmap
# Shows relationships between numeric variables
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title('HeatMap Chart')
plt.show()
```
